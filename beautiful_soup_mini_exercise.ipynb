```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import re
```

```python
response = requests.get('https://web-scraping-demo.zgulde.net/news')
soup = BeautifulSoup(response.text)
articles = soup.select('.grid.gap-y-12 > div')
```

```python
def parse_news_article(article):
    title = article.h2.text
    date, author = article.select('.italic')[0].find_all('p')
    summary = article.select('p')[-1]
    return {'title': title, 'date': date.text, 'author': author.text, 'summary':summary.text}
```

```python
pd.DataFrame([parse_news_article(article) for article in articles])
```

```python
def parse_contacts(contact):
    name = contact.h2.text
    quote = contact.p.text.strip()
    email = contact.select('.email')[0].text
    address = contact.select('.address')[0].text.strip()
    address = re.sub(r'\s{2,}',' ', address)
    return {'name': name, 'quote': quote, 'email': email, 'address':address}
```

```python
parse_contacts(contact)
```

```python
pd.DataFrame([parse_contacts(contact) for contact in contacts])
```
